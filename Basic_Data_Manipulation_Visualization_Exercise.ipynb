###  28. Import the dataset occupation.csv and assign object as users

### 29. Discover what is the mean age per occupation

### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
import pandas as pd
import numpy as np

# Import the dataset with pipe separator and assign it to an object called users
#users = pd.read_csv('occupation.csv', sep='|')

# Encode the gender column: Male = 1, Female = 0
users['gender_encoded'] = np.where(users['gender'] == 'M', 1, 0)

# Calculate the male ratio per occupation
# First, calculate the total number of individuals and the number of males per occupation
gender_ratio = users.groupby('occupation').agg(
    total_count=('gender_encoded', 'size'),
    male_count=('gender_encoded', 'sum')
)


# Calculate the male ratio
gender_ratio['male_ratio'] = gender_ratio['male_count'] / gender_ratio['total_count']

# Sort the results from most to least male ratio
sorted_gender_ratio = gender_ratio.sort_values(by='male_ratio', ascending=False)

# Display the result
print(sorted_gender_ratio[['male_ratio']])
```

### 31. For each occupation, calculate the minimum and maximum ages

```python

# Calculate the minimum and maximum ages per occupation
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Display the result
print(age_range_per_occupation)
```

### 32. For each combination of occupation and gender, calculate the mean age

```python

# Calculate the mean age for each combination of occupation and gender
mean_age_by_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean().reset_index()

# Display the result
print(mean_age_by_occupation_gender)
```

### 33.  For each occupation present the percentage of women and men

```python


# Encode the gender column: Male = 1, Female = 0
users['gender_encoded'] = np.where(users['gender'] == 'M', 1, 0)

# Calculate the total number of individuals, and number of women and men per occupation
gender_counts = users.groupby('occupation').agg(
    total_count=('gender_encoded', 'size'),
    male_count=('gender_encoded', 'sum'),
    female_count=('gender_encoded', lambda x: (x == 0).sum())
)

# Calculate the percentage of women and men
gender_counts['percentage_male'] = (gender_counts['male_count'] / gender_counts['total_count']) * 100
gender_counts['percentage_female'] = (gender_counts['female_count'] / gender_counts['total_count']) * 100

# Drop intermediate count columns
gender_percentage = gender_counts[['percentage_male', 'percentage_female']]

# Display the result
print(gender_percentage)
```

### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
from datetime import datetime

# Get the current date and time
now = datetime.now()

# a. Current date and time
current_date_time = now
print("a. Current date and time:", current_date_time)

# b. Current year
current_year = now.year
print("b. Current year:", current_year)

# c. Month of year
current_month = now.month
print("c. Month of year:", current_month)

# d. Week number of the year
week_number = now.isocalendar()[1]
print("d. Week number of the year:", week_number)

# e. Weekday of the week (0=Monday, 6=Sunday)
weekday = now.weekday()
weekday_name = now.strftime("%A")  # Full weekday name
print("e. Weekday of the week (number):", weekday)
print("   Weekday of the week (name):", weekday_name)

# f. Day of year
day_of_year = now.timetuple().tm_yday
print("f. Day of year:", day_of_year)

# g. Day of the month
day_of_month = now.day
print("g. Day of the month:", day_of_month)

# h. Day of week (e.g., Monday, Tuesday, etc.)
day_of_week = now.strftime("%A")
print("h. Day of week:", day_of_week)
```

### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM

Expected Output : 2014-07-01 14:43:00

```python
from datetime import datetime

# Sample string
date_string = "Jul 1 2014 2:43PM"

# Convert the string to a datetime object
date_format = "%b %d %Y %I:%M%p"  # Define the format
date_object = datetime.strptime(date_string, date_format)

# Print the result in the desired format
formatted_date = date_object.strftime("%Y-%m-%d %H:%M:%S")
print("Converted datetime:", formatted_date)
```

### 52. Write a Python program to convert unix timestamp string to readable date.

Sample Unix timestamp string : 1284105682
    
Expected Output : 2010-09-10 13:31:22

```python
from datetime import datetime

# Sample Unix timestamp string
timestamp_str = "1284105682"

# Convert the timestamp string to an integer
timestamp = int(timestamp_str)

# Convert the Unix timestamp to a datetime object
date_time = datetime.fromtimestamp(timestamp)

# Format the datetime object as a readable string
formatted_date = date_time.strftime("%Y-%m-%d %H:%M:%S")

# Print the result
print("Readable Date :", formatted_date)
```

### 89. Create line plots for Soap & Facewash of all months in a single plot using Subplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
company_data = pd.read_csv('FMCG_Company_Data_2019.csv')

# Set the 'Month' column as the index (assuming it exists and is correctly formatted)
company_data['Month'] = pd.to_datetime(company_data['Month'], format='%b-%y')  # Adjust format as needed
company_data.set_index('Month', inplace=True)

# Extract data for 'Soap' and 'Facewash'
soap_data = company_data['Soap']
facewash_data = company_data['FaceWash']

# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plot 'Soap'
ax[0].plot(soap_data.index, soap_data, marker='o', linestyle='-', color='blue', label='Soap')
ax[0].set_title('Monthly Sales of Soap')
ax[0].set_ylabel('Units Sold')
ax[0].legend()
ax[0].grid(True)

# Plot 'Facewash'
ax[1].plot(facewash_data.index, facewash_data, marker='o', linestyle='-', color='green', label='Facewash')
ax[1].set_title('Monthly Sales of Facewash')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Units Sold')
ax[1].legend()
ax[1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### 90. Create Box Plot for Total Profit variable

```python
import pandas as pd

# Load the dataset
company_data = pd.read_csv('FMCG_Company_Data_2019.csv')

# Inspect the dataset to confirm the 'Total Profit' column
print(company_data.head())

# Check the data type of 'Total Profit'
print(company_data['Total_Profit'].dtype)

# Convert to numeric if it's not already
company_data['Total_Profit'] = pd.to_numeric(company_data['Total_Profit'], errors='coerce')

import matplotlib.pyplot as plt

# Create a box plot for 'Total Profit'
plt.figure(figsize=(10, 6))
plt.boxplot(company_data['Total_Profit'].dropna(), vert=False)
plt.title('Box Plot of Total Profit')
plt.xlabel('Total Profit')
plt.grid(True)
plt.show()
```
