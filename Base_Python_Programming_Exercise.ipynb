```python
# Python program to find numbers divisible by 7 but not a multiple of 5

# List to store the numbers
numbers = []

# Loop through the range 2000 to 3200 (inclusive)
for i in range(2000, 3201):
    if i % 7 == 0 and i % 5 != 0:
        numbers.append(str(i))

# Join the list into a comma-separated string and print it
print(", ".join(numbers))
```

```python
# Python program to count elements of each data type in a list

# Given list
data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

# Dictionary to store the count of each data type
type_counts = {}

# Loop through each element in the list
for element in data:
    # Get the data type of the element
    element_type = type(element)

    # If the type is already in the dictionary, increment the count
    if element_type in type_counts:
        type_counts[element_type] += 1
    # Otherwise, add the type to the dictionary with a count of 1
    else:
        type_counts[element_type] = 1

# Display the results
for element_type, count in type_counts.items():
    print(f"{element_type.__name__}: {count}")
```

```python
# Python program to add all numeric values from the list

# Given list
data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

# Variable to store the sum of numeric values
total_sum = 0

# Loop through each element in the list
for element in data:
    # Check if the element is of numeric type (int, float, bool)
    if isinstance(element, (int, float)):
        total_sum += element

# Display the result
print(f"The sum of all numeric values is: {total_sum}")
```

```python
# Python program to concatenate all string values with a hyphen as a delimiter

# Given list
data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

# List to store all string values
string_values = []

# Loop through each element in the list
for element in data:
    # Check if the element is a string
    if isinstance(element, str):
        string_values.append(element)

# Concatenate the strings with a hyphen as a delimiter
result = "-".join(string_values)

# Display the result
print(f"The concatenated string is: {result}")
```

```python
# Define the UDF
def my_func(data):
    # Initialize variables
    total_sum = 0
    str_count = 0

    # Loop through each element in the list
    for element in data:
        # Check if the element is a number (int, float) but not a bool
        if isinstance(element, (int, float)) and not isinstance(element, bool):
            total_sum += element
        # Check if the element is a string
        elif isinstance(element, str):
            str_count += 1

    # Return the results as a dictionary
    return {'Sum': total_sum, 'Count_of_Strs': str_count}

# Example usage
l1 = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]
result = my_func(l1)
print(result)
```

```python
# Using a loop to get odd numbers
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

odd_numbers_loop = []

for num in li:
    if num % 2 != 0:
        odd_numbers_loop.append(num)

print("Odd numbers using loop:", odd_numbers_loop)


# Using list comprehensions to get odd numbers
odd_numbers_list_comp = [num for num in li if num % 2 != 0]

print("Odd numbers using list comprehension:", odd_numbers_list_comp)


# Using lambda function with filter to get odd numbers
odd_numbers_lambda = list(filter(lambda x: x % 2 != 0, li))

print("Odd numbers using lambda with filter:", odd_numbers_lambda)
```

```python
# Define the UDF to calculate descriptives
def calculate_descriptives(numbers):
    # Calculate sum
    total_sum = sum(numbers)

    # Calculate count
    count = len(numbers)

    # Calculate minimum
    minimum = min(numbers)

    # Calculate mean
    mean = total_sum / count if count > 0 else 0

    # Calculate maximum
    maximum = max(numbers)

    # Return the descriptives as a dictionary
    return {
        'Sum': total_sum,
        'Count': count,
        'Min': minimum,
        'Mean': mean,
        'Max': maximum
    }

# Example usage
numbers_list = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]
descriptives = calculate_descriptives(numbers_list)
print(descriptives)
```

```python
import math

# Define the UDF to calculate the area of different shapes
def calculate_area(shape, **kwargs):
    shape = shape.lower()

    if shape == 'square':
        # Validate that 'side' is provided
        if 'side' in kwargs:
            side = kwargs['side']
            if isinstance(side, (int, float)) and side > 0:
                return side * side
            else:
                return "Please enter a positive numeric value for 'side'."
        else:
            return "Please enter 'side' for a square."

    elif shape == 'rectangle':
        # Validate that 'length' and 'width' are provided
        if 'length' in kwargs and 'width' in kwargs:
            length = kwargs['length']
            width = kwargs['width']
            if all(isinstance(dim, (int, float)) and dim > 0 for dim in [length, width]):
                return length * width
            else:
                return "Please enter positive numeric values for 'length' and 'width'."
        else:
            return "Please enter both 'length' and 'width' for a rectangle."

    elif shape == 'circle':
        # Validate that 'radius' is provided
        if 'radius' in kwargs:
            radius = kwargs['radius']
            if isinstance(radius, (int, float)) and radius > 0:
                return math.pi * radius * radius
            else:
                return "Please enter a positive numeric value for 'radius'."
        else:
            return "Please enter 'radius' for a circle."

    else:
        return "Invalid shape. Please choose either 'square', 'rectangle', or 'circle'."

# Example usage
print(calculate_area('square', side=4))
print(calculate_area('rectangle', length=5, width=3))
print(calculate_area('circle', radius=7))
print(calculate_area('square'))  # Error case
print(calculate_area('rectangle', length=5))  # Error case
print(calculate_area('circle', diameter=7))  # Error case
print(calculate_area('triangle', base=5, height=3))  # Invalid shape
```

```python
def reconcile_lists(l1, l2):
    # Convert lists to sets for easier set operations
    set1 = set(l1)
    set2 = set(l2)

    # Find matched values (intersection)
    matched = list(set1.intersection(set2))

    # Find values only in l1 (difference)
    only_in_l1 = list(set1.difference(set2))

    # Find values only in l2 (difference)
    only_in_l2 = list(set2.difference(set1))

    # Return results as a dictionary
    return {
        'Matched': sorted(matched),
        'Only in l1': sorted(only_in_l1),
        'Only in l2': sorted(only_in_l2)
    }

# Example usage
l1 = ['January', 'February', 'March', 'May', 'June', 'September', 'December']
l2 = ['January', 'February', 'April', 'June', 'October', 'December']
result = reconcile_lists(l1, l2)
print(result)
```

```python
def is_prime(number):
    # Check if the number is less than 2 (not prime)
    if number <= 1:
        return False

    # Check if the number is 2 or 3 (both prime)
    if number <= 3:
        return True

    # Eliminate multiples of 2 and 3
    if number % 2 == 0 or number % 3 == 0:
        return False

    # Check divisibility from 5 to sqrt(number)
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6

    return True

# Example usage
print(is_prime(2))    # True
print(is_prime(4))    # False
print(is_prime(29))   # True
print(is_prime(100))  # False
```

```python
import math

# Function to compute the factorial
def compute_factorial(n):
    return math.factorial(n)

# Function to process a list of inputs
def process_factorials(inputs):
    results = []
    for value in inputs:
        try:
            number = int(value)
            if number < 0:
                results.append(f"Factorial is not defined for negative numbers.")
            else:
                result = compute_factorial(number)
                results.append(result)
        except ValueError:
            results.append("Invalid input. Please enter a valid integer.")
    return results

# Get user input
user_input = input("Enter numbers separated by commas: ")

# Split the input into a list of strings
input_list = user_input.split(',')

# Process the list of inputs
factorial_results = process_factorials(input_list)

# Print the results in a comma-separated sequence
print(', '.join(map(str, factorial_results)))
```

```python
def generate_square_dict(n):
    # Generate a dictionary where keys are integers and values are their squares
    square_dict = {i: i * i for i in range(1, n + 1)}
    return square_dict

# Get user input
try:
    # Read input from the user
    user_input = int(input("Enter an integer: "))

    # Check if the input is a positive integer
    if user_input < 1:
        print("Please enter a positive integer greater than 0.")
    else:
        # Generate and print the dictionary
        result = generate_square_dict(user_input)
        print(result)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def generate_list_and_tuple(input_string):
    # Split the input string into a list of numbers
    number_list = input_string.split(',')

    # Convert the list to a tuple
    number_tuple = tuple(number_list)

    return number_list, number_tuple

# Get user input
user_input = input("Enter a sequence of comma-separated numbers: ")

# Generate list and tuple
list_result, tuple_result = generate_list_and_tuple(user_input)

# Print the results
print(list_result)
print(tuple_result)
```

```python
def sort_words_alphabetically(input_string):
    # Split the input string into a list of words
    words_list = input_string.split(',')

    # Sort the list of words alphabetically
    sorted_words = sorted(words_list)

    # Join the sorted list into a comma-separated string
    sorted_words_string = ','.join(sorted_words)

    return sorted_words_string

# Get user input
user_input = input("Enter a sequence of comma-separated words: ")

# Process and print the sorted words
sorted_words = sort_words_alphabetically(user_input)
print(sorted_words)
```

```python
def process_words(input_string):
    # Split the input string into a list of words
    words_list = input_string.split()

    # Remove duplicate words by converting the list to a set
    unique_words = set(words_list)

    # Sort the unique words alphabetically
    sorted_words = sorted(unique_words)

    # Join the sorted words into a whitespace-separated string
    sorted_words_string = ' '.join(sorted_words)

    return sorted_words_string

# Get user input
user_input = input("Enter a sequence of whitespace-separated words: ")

# Process and print the result
result = process_words(user_input)
print(result)
```

```python
def count_case_letters(sentence):
    # Initialize counters for uppercase and lowercase letters
    upper_case_count = 0
    lower_case_count = 0

    # Iterate through each character in the sentence
    for char in sentence:
        if char.isupper():
            upper_case_count += 1
        elif char.islower():
            lower_case_count += 1

    return upper_case_count, lower_case_count

# Get user input
user_input = input("Enter a sentence: ")

# Count uppercase and lowercase letters
upper_count, lower_count = count_case_letters(user_input)

# Print the results
print(f"UPPER CASE {upper_count} LOWER CASE {lower_count}")
```

```python
def reverse_string(s):
    # Reverse the string using slicing
    return s[::-1]

# Get user input
user_input = input("Enter a string: ")

# Reverse the string
reversed_string = reverse_string(user_input)

# Print the reversed string
print(reversed_string)
```
